#### Use plt.subplots(nrows=1, ncols=2) to create the plot.

```python
# Method 2

fig, axes = plt.subplots(ncols=2, nrows=1) 

axes[0].set_title('First Plot')
axes[0].plot(x, y)

axes[1].set_title('Second Plot')
axes[1].plot(x,z)


plt.show();
```

#### Use your previous code but now, resize your plot.
**Hint**: Add the figsize argument in plt.subplots().

If you want, try to add a title to the plot or even axes labels. You can also play with the fontweight and fontsize of the titles and labels. 

```python
# your code here

fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(8, 4))

axes[0].set_title('First Plot')
axes[0].plot(x, y,label='x vs y')
axes[0].xaxis.set_tick_params(rotation=45)
axes[0].legend()

axes[1].set_title('Second Plot')
axes[1].plot(x, z, label='x vs z')
axes[1].xaxis.set_tick_params(rotation=45)
axes[1].legend()

plt.show()
```

#### Now you will have to plot both $y=x^2$ and $y=e^x$ in the same plot. Create two subplots and use a normal scale in the first one and a logarithmic scale in the second one.
**Hint**: Use `set_xscale` and `set_yscale`.

```python
# your code here

fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(8, 4))

axes[0].set_title('Linear Plot')
axes[0].plot(x,x**2)
axes[0].plot(x,np.exp(x))
axes[0].set_yscale('linear')


axes[1].set_title('Log Plot')
axes[1].plot(x,x**2)
axes[1].plot(x,np.exp(x))
axes[1].set_yscale('log')


plt.show()
```

#### As a bonus challenge, try to add a legend to the plot.

```python
# your code here

fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(8, 4))

axes[0].set_title('Linear Plot')
axes[0].plot(x,x**2,label='x vs x**2')
axes[0].plot(x,np.exp(x),label='x vs np.exp(x)')
axes[0].set_yscale('linear')
axes[0].legend()

axes[1].set_title('Log Plot')
axes[1].plot(x,x**2,label='x vs x**2')
axes[1].plot(x,np.exp(x),label='x vs np.exp(x)')
axes[1].set_yscale('log')
axes[1].legend()

plt.show()
```
